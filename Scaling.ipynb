```python
import numpy as np
# 로그 변환 전
plt.hist(new_titanic_df['fare'])
plt.show()

# 로그 변환 후
Log_Fare = np.log1p(new_titanic_df['fare'])
plt.hist(Log_Fare)
plt.show()
```

```python
np.expand_dims(np.array(new_titanic_df['age']),0)
```

```python
# minmax scaler
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(np.expand_dims(np.array(new_titanic_df['age']),1))
scaled = scaler.transform(np.expand_dims(np.array(new_titanic_df['age']),1))
```

```python
plt.hist(new_titanic_df['age'])
plt.show()

plt.hist(scaled.reshape(-1))
plt.show()
```
