# 시각화 예제

```python
# 시각화 기본 함수
import matplotlib.pyplot as plt

# 기본 그래프 그리기
X = [1,2,3,4]
Y = [2,4,6,8]

plt.plot(X, Y)
plt.show()
```

```python
# 제목, 축 레이블,범례 설정

X = [1,2,3,4]
Y = [2,4,6,8]

plt.title('y=2x plot')
plt.plot(X, Y, label='y=2x')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

# Histogram

```python
# 붓꽃 데이터 셋 - 붓꽃의 품종 판별
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

iris_df = sns.load_dataset('iris')
```

```python
iris_df.head(3)
```

```python
plt.hist(iris_df['petal_length'])
plt.show()
```

```python
import matplotlib.pyplot as plt

# grouping
A = iris_df.loc[iris_df['species'] == 'setosa', 'petal_length']
B = iris_df.loc[iris_df['species'] == 'versicolor', 'petal_length']
C = iris_df.loc[iris_df['species'] == 'virginica', 'petal_length']

# 레이블 별로 히스토그램 생성
plt.hist(A, alpha=0.5, label='Label 0')
plt.hist(B, alpha=0.5, label='Label 1')
plt.hist(C, alpha=0.5, label='Label 2')

plt.legend() # Label 확인
plt.show()
```
