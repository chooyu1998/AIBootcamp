# 리스트

- [] 로 선언
- 순서가 존재함
- 추가/변경/삭제 가능
- 중복값 가능

```python
# 리스트 생성

a = []                                              # 빈 리스트 생성
b = [1,2,3]
c = ['AI','bootcamp','Python','Programming']
d = [2023,'AI','bootcamp']                          # 자료형을 섞어서 리스트를 선언할 수 있음
e = [i for i in range(0,6)]                         # for문으로 선언 가능 (0부터 5까지 정수 생성)
f = [1,2,[1,2,3]]                                   # 이중 배열 생성

print(a)
print(b)
print(c)
print(d)
print(e)
print(f)
print(' ')

# 길이 확인
print(len(a))
print(len(b))
print(len(c))
print(len(d))
print(len(e))
print(len(f))
```

* 리스트 정렬
  - sort : 역순 정렬은 reverse=True 추가

```python
li = [5,2,3,1,10,8]

print('sort before', li)
sorted_li = sorted(li)
reverse_sorted_li = sorted(li, reverse = True)
print('sort after', sorted_li)
print('sort after(reverse)', reverse_sorted_li)

```

* 값 추가
  - insert : 정해진 위치에 삽입 가능
  - append : 맨뒤에 추가

```python
# insert 설명
li = ['AI','bootcamp','Programming']

print('insert before', li)
li.insert(2,'Python') # 세번째 위치(2) 에 'Python' 삽입
print('insert after', li)

# append 설명

li = ['AI','bootcamp','Python']
print('append before',li)
li.append('Programming')
print('append after', li)
```

* 값 삭제
  - pop : 값 추출 후 삭제
  - remove : 값 추출 없이 삭제

```python
li = ['AI','bootcamp','Python','Programming']

print('pop before ', li)
print('pop ', li.pop(2)) # 추출하고 싶은 값의 인덱스를 지정
print('pop after', li)
print(' ')

li = ['AI','bootcamp','Python','Programming','Python']
print('remove before', li)
li.remove('Python') # 추출하고 싶은 값을 지정 (단 첫번째 위치만 삭제함)
print('remove after', li)
```

# 딕셔너리

* key와 value의 쌍을 갖는 자료형

```python
dictionary = { # 딕셔너리 선언 (이름:나이)
    '홍길동' : 23,
    '김종현' : 24,
    '이동준' : 30,
    '송하윤' : 19,
    '정예원' : 27,
    '홍길동' : 34
}


print(dictionary)
#print(dictionary.keys()) # 키 값 출력
#print(dictionary.values()) # value 값 출력
#print(dictionary.items()) # 전체 출력
#print(dictionary['홍길동']) # 홍길동의 나이 출력
#print(' ')

# 예제
#for key,value in dictionary.items():
#  print(key,'의 나이는', value,'입니다.')
```

* 딕셔너리 값 추가/삭제/변경
  - 삭제 : del과 pop이 있음
    - del : 값 추출 없이 삭제
    - pop : 값 추출 후 삭제

```python
dictionary = {
    '홍길동' : 23,
    '김종현' : 24,
    '이동준' : 30,
    '송하윤' : 19,
    '정예원' : 27
} # 딕셔너리 선언 (이름:나이)

# 값 추가
print('before', dictionary)
dictionary['장서연'] = 26
print('after', dictionary)
print(' ')

# del을 통한 값 삭제
print('before', dictionary)
del dictionary['홍길동']
print('after', dictionary)
print(' ')

#pop을 통한 값 삭제
print('before',dictionary)
print('pop', dictionary.pop('정예원'))
print('after', dictionary)
print(' ')

# 값 변경
print('before', dictionary)
dictionary['이동준'] = 34
print('after', dictionary)
```

# 집합

* 순서가 없음, 따라서 인덱싱/슬라이싱 불가
* 중복을 허용하지 않음
* 집합 연산 가능 (교집합, 합집합, 차집합)

```python
x_set = {1,2,4,5,6}

print(x_set)
print(len(x_set))
# print(x_set[0]) 인덱싱 불가
```

* 집합 값 추가, 단 중복 값일 경우 무시
  - add : 원하는 값 추가 가능,
  - update : 원하는 값 여러개 추가 가능
* 집합 값 삭제 (다른 자료형과 동일)
   - remove : 특정 값 추출 없이 제거
   - pop : 특정 값 추출 후 제거

```python
x_set = {1,2,3,4}
# 값 추가 하는 방법

# add
print('before ', x_set)
x_set.add(5)
print('after ', x_set)

# update
print('before ', x_set)
x_set.update({7,9,10,1,2,8})
print('after ', x_set)
```

* 집합 연산자

```python
set_1 = set([1,2,3,4,5])
set_2 = set([4,5,6,7,8])

# 교집합
print(set_1.intersection(set_2))
print(set_1 & set_2)

# 합집합
print(set_1.union(set_2))
print(set_1 | set_2)

# 차집합
print(set_1.difference(set_2))
print(set_1 - set_2)
```

# Numpy

* Numerical Python으로 벡터 및 행렬 연산에 있어서 편리한 기능을 제공함

* numpy 라이브러리를 import 해줘야함

```python
import numpy as np
```

* 배열의 생성

```python
# 리스트를 통한 array 정의
data_1 = [1,2,3,4,5]
arr_1 = np.array(data_1) #1차원 행렬
print('array 1')
print(arr_1)

data_2 = [[1,2,3],[4,5,6]]
arr_2 = np.array(data_2) #2차원 행렬
print('array 2')
print(arr_2)

# 배열의 차원 확인

print('array 1 shape', arr_1.shape)
print('array 2 shape', arr_2.shape) #2행 3열
```

* 특수한 배열의 생성
   - arange 함수
   - zeros 함수
   - ones 함수

```python
# arange
arange_arr_1 = np.arange(10) # 0부터 1씩 증가하는 1차원 배열 (9까지)
print('arange 1')
print(arange_arr_1)
arange_arr_2 = np.arange(5,10) #5부터 1씩 증가하는 1차원 배열
print('arange 2')
print(arange_arr_2)


# zeros
zeros_arr_1 = np.zeros(10) # 길이가 10인 영행렬 생성
print('zeros 1')
print(zeros_arr_1)
zeros_arr_2 = np.zeros((2,2)) # 2x2의 영행렬 생성
print('zeros 2')
print(zeros_arr_2)

# ones
ones_arr_1 = np.ones(10) # 길이가 10인 유닛행렬 생성
print('ones 1')
print(ones_arr_1)
ones_arr_2 = np.ones((5,3)) # 5x3의 유닛행렬 생성
print('ones 2')
print(ones_arr_2)

```

##  배열간의 연산

* +,-,*,/ 등의 연산자 사용이 가능함

```python
arr_1 = np.array([1,2,3])
arr_2 = np.array([4,5,6])

# 행렬 더하기, 각 행/열에 있는 값끼리 더해짐
sum_arr = arr_1 + arr_2
print('sum ', sum_arr)

# 행렬 빼기
sub_arr = arr_1 - arr_2
print('subtract', sub_arr)

# 행렬 곱하기
multi_arr = arr_1 * arr_2
print('multiply', multi_arr)

# 행렬 나누기
divide_arr = arr_1 / arr_2
print('divide', divide_arr)
```

* 행렬 곱/내적 연산 (2차원에서는 동일)
  - dot : 뒤에 실습내용 보고 추가할지 결정

```python
arr_1 = np.array([[1,2],[3,4]])
arr_2 = np.array([[5,6],[7,8]])

dot_arr = np.dot(arr_1, arr_2)
print(dot_arr)

arr_1 = np.array([[1,2,3],[4,5,6]])
arr_2 = np.array([[6,23],[-1,7],[8,9]])

# 2x3 * 3x2 = 2x2
dot_arr = np.dot(arr_1, arr_2)
print(dot_arr)
```

* 모든 원소에 대한 계산
 - sum() : 모든 원소의 합
 - prod() : 모든 원소의 곱
 - mean() : 모든 원소의 평균
 - median() : 모든 원소의 중앙값
 - std() : 모든 원소의 표준편차
 - var() : 모든 원소의 분산

```python
# 1차원 배열 예시
arr = np.array([1,2,3,4,5,6])

sum = np.sum(arr)
print('sum', sum)

prod = np.prod(arr)
print('prod', prod)

mean = np.mean(arr)
print('mean', mean)

median = np.median(arr)
print('median',median)

std = np.std(arr)
print('std', std)

var = np.var(arr)
print('var', var)

# 2차원 배열이면 ?
arr = np.array([[1,2,3],
                [4,5,6]])

# 행별/열별로 계산이 가능함
print('sum by row', np.sum(arr, axis=0)) # sum by row
print('sum by col', np.sum(arr, axis=1)) # sum by column
print('total sum', np.sum(arr)) # total sum
```


 - 리스트처럼 슬라이싱하여 추출 가능
 - loc : 지정해놓은 키 값으로 접근 가능
 - iloc : 인덱스로만 접근 가능

```python
df = pd.DataFrame({
    'Name' : ['홍길동','김종현','이동준','송하윤','정예원'],
    'Age' : [23,24,30,19,27],
    'City' : ['서울','부산','인천','부산','대구']
})


print(df[0:3]) # 0부터 2까지의 행 데이터 추출
print(' ')
print(df.loc[0:2, ['Name','Age']]) # 키 값 사용 가능
print(' ')
print(df.iloc[0:2, 0:2]) # 인덱스 값으로만 사용 가능
#print(df.iloc[0:2,['Name']]) <- 이렇게 하면 오류가 발생함
```

* 조건문을 통한 데이터 추출

```python
df = pd.DataFrame({
    'Name' : ['홍길동','김종현','이동준','송하윤','정예원','이종윤'],
    'Age' : [23,24,30,19,27,26],
    'City' : ['서울','부산','인천','부산','대구','부산']
})

print(df[df['City'] == '부산']) # 부산에 거주하는 사람들의 데이터 추출
print(df[df['Age'] < 25]) # 나이가 25 미만인 사람들의 데이터 추출
print(df[ (df['City'] == '부산') & (df['Age'] < 25) ]) # 여러개의 조건 사용 가능
```

# 시각화 예제

```python
# 시각화 기본 함수
import matplotlib.pyplot as plt

# 기본 그래프 그리기
X = [1,2,3,4]
Y = [2,4,6,8]

plt.plot(X, Y)
plt.show()
```

```python
# 제목, 축 레이블 설정

X = [1,2,3,4]
Y = [2,4,6,8]

plt.title('y=2x plot')
plt.plot(X, Y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# 범례 표시하기

X1 = [1,2,3,4]
Y1 = [2,4,6,8]
X2 = [1,2,3,4]
Y2 = [1,2,3,4]

plt.plot(X1, Y1, label='y=2x')
plt.plot(X2,Y2, label='y=x')
plt.legend()
plt.show()
```

## Histogram

```python
# 붓꽃 데이터 셋 - 붓꽃의 품종 판별
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

iris_df = sns.load_dataset('iris')
```

```python
iris_df.head(3)
```

```python
plt.hist(iris_df['petal_length'])
plt.show()
```

```python
import matplotlib.pyplot as plt

# grouping
A = iris_df.loc[iris_df['species'] == 'setosa', 'petal_length']
B = iris_df.loc[iris_df['species'] == 'versicolor', 'petal_length']
C = iris_df.loc[iris_df['species'] == 'virginica', 'petal_length']

# 레이블 별로 히스토그램 생성
plt.hist(A, alpha=0.5, label='Label 0')
plt.hist(B, alpha=0.5, label='Label 1')
plt.hist(C, alpha=0.5, label='Label 2')

plt.legend() # Label 확인
plt.show()
```
